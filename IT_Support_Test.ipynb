```python
import requests
import json

url = "https://siri-chandana.app.n8n.cloud/webhook/it-support"

test_cases = [
    {"user": "John", "issue": "I forgot my password and cannot login"},
    {"user": "Sarah", "issue": "My WiFi keeps disconnecting"},
    {"user": "Mike", "issue": "Microsoft Teams crashes when I share screen"}
]

for case in test_cases:
    print(f"Reporter : {case['user']}")
    print(f"Issue    : {case['issue']}")

    response = requests.post(url, json=case)

    raw = response.text.lstrip("=")
    outer = json.loads(raw)
    result = json.loads(outer["text"])

    print(f"Category : {result['category']}")
    print(f"Priority : {result['priority']}")
    print(f"Solution : {result['solution']}")
    print(f"Escalate : {result['escalate']}")
    print("-" * 40)
```
